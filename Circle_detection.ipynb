```python
while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Convert the frame to grayscale
        gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY) 
        
        # Blur using 3x3 kernel
        gray_blurred = cv.blur(gray, (3, 3)) 
        
        # Apply Hough Circle Transform
        detected_circles = cv.HoughCircles(gray_blurred, cv.HOUGH_GRADIENT, 1, 20, 
                                           param1=50, param2=30, minRadius=1, maxRadius=40) 
        
        # Draw circles that are detected
        if detected_circles is not None:
            # Convert the circle parameters to integers
            detected_circles = np.uint16(np.around(detected_circles)) 
            
            for pt in detected_circles[0, :]: 
                a, b, r = pt[0], pt[1], pt[2] 

                # Draw the circumference of the circle
                cv.circle(frame, (a, b), r, (0, 255, 0), 2) 

                # Draw a small circle to show the center
                cv.circle(frame, (a, b), 1, (0, 0, 255), 3) 
            
        cv.imshow("Detected Circles", frame)
        
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
```

```python
cap.release()
cv.destroyAllWindows()
```
